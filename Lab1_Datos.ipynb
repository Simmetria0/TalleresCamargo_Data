```python
####################################################
#  4. Control Flow
####################################################
# creo una variable
some_var = 5

# Ejemplo de IF
if some_var > 10:
    print ("some_var es mas grande de 10.")
```

```python
#IF flow mas completo:
if some_var > 10:
    print ("some_var es mas grande de 10.")
elif some_var < 10:  
    print ("some_var es minor de 10.")
else:  
    print ("some_var es 10.")
```

```python
# y WHILE que iteran hasta que una coindicion sea verdadera
x = 0
while x < 4:
    print (x)
    x += 1  
```

```python
import numpy as np

#vamos a crear un vector
a = np.array([1, 2, 3])  
```

```python
#=====================
#explorar los datos:
#=====================
print(type(a))          

print(a.shape)           
```

```python
#acceso a elementos
print(a[0])  

#cambia valore de elmentos
a[0] = 5
```

```python
#MATRIXES
b = np.array([[1,2,3],[4,5,6]])    
print(b.shape)                     
```

```python
#Numpy tiene muchas maneras para crear vectores y matrices
x = np.arange(7)
y = np.arange(3,7)

# Crea una matriz de zeros
a = np.zeros((2,2))  
print(a)              
```

```python
# ...de unos
b = np.ones((1,2))    
print(b)             

#con una costante
c = np.full((2,2), 7)  
print(c)               
                     

# Matriz identidad    
d = np.eye(2)        
print(d)              
                     
# de numeros casuales
e = np.random.random((2,2)) 
print(e)                    
```
