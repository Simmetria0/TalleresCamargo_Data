# URLLIB



```python
#############
# ACTION
from urllib import request

dir(request)
```

```python
resp = request.urlopen('https://www.wikipedia.org')
```

```python
#Averiguamos si la request ha tenido exito
resp.code
```


---
### 200 --> OK
### 400 ---> Bad request
### 403 ---> Forbidden
### 404 ---> Not found
---


```python
#size of the response in bytes
resp.length
```

```python
#para ver solo parte de la response
resp.peek()
```

```python
#vamos a verlo 
data = resp.read()
type(data)
```

```python
len(data)
```

```python
#son bytes, para reconvertirlo en html, especifica el encode
html = data.decode("UTF-8")
```

```python
#ahora..
type(html)
```
